```python
import nltk
from nltk.chat.util import Chat ,reflections

nltk.download('punkt')

#defining the pairs
pairs=[
    [
        r"my name is (.*)",
        ["Hello %1 ,How are you Today?",]
    ],
    [
        r"hi|hey|hello",
        ["Hello %1","How Can I Help You",]
    ],
    [
        r"what is your name?",
        ["You can call me as Olivia , Because personally i like that name!!", ]
    ],
        [
            r"who introduced you?",
            ["I m introduced by Miss.Shruti Ganpat Waghmare in year 2024",]
        ],
        [
            r"how are you?",
            ["I m good ! , What About You?",]
        ],
        [
            r"sorry",
            ["It's Alright , No Issue",]
        ],
        [
            r"do you know cooking?",
            ["Yes , I like it !",]
        ],
        [
            r"give me recipe of mohito|mojito|moeto",
            ['''Certainly! Here's a short and simple recipe for a classic Mojito:
- 10 fresh mint leaves
- 1/2 lime, cut into 4 wedges
- 2 tablespoons white sugar (to taste)
- 1 cup ice cubes
- 1 1/2 ounces white rum
- 1/2 cup club soda

 Instructions:
1.Muddle Mint and Lime: Place mint leaves and 1 lime wedge into a sturdy glass. Use a muddler to crush the mint and lime to release the mint oils and lime juice.
2. Add Sugar: Add 2 more lime wedges and the sugar, and muddle again to release the lime juice. Do not strain the mixture.
3. Add Ice and Rum: Fill the glass almost to the top with ice. Pour the rum over the ice.
4. Top with Club Soda: Add club soda to the top and stir well to combine.
5. Garnish and Serve: Garnish with the remaining lime wedge and a sprig of mint. Serve immediately.

Enjoy your refreshing Mojito!''',]
        ],
        [
            r"give me recipe of mango juice",
            ['''Here's a simple and quick recipe for refreshing mango juice:

 Mango Juice Recipe

 Ingredients:
- 2 ripe mangoes
- 1 cup cold water (or adjust for desired consistency)
- 1-2 tablespoons sugar (optional, adjust to taste)
- 1-2 teaspoons lemon juice (optional, for a tangy flavor)
- Ice cubes (optional)

Instructions:
1. Prepare the Mangoes: Peel the mangoes and cut the flesh away from the pit. Chop the mango flesh into small pieces.
2. Blend: Place the mango pieces in a blender. Add the cold water and sugar (if using). Blend until smooth.
3. Adjust Consistency: If the juice is too thick, add a little more water and blend again until you reach the desired consistency.
4. Add Lemon Juice: Add the lemon juice if you like a tangy flavor. Blend briefly to mix.
5. Serve: Pour the mango juice into glasses over ice cubes if desired.

Enjoy your homemade mango juice!''',]
        ],
        [
            r"quit",
            ["By Take Care ! , It was Nice to Talk With You .",]
        ],
        [
            r"(.*)",
            ["I am not able to understand , please can you rephrase?",]
        ],
]
        
#Basic Reflections
reflections = {
            "i am" : "you are",
            "i was" : "you were",
            "i" : "you",
            "i'd" : "you would",
            "i'll" : "you will",
            "i've" : "you have",
            "my" : "your",
            "you are" : "i am",
            "you were" : " i was",
            "you've" : "i have",
            "you'll" : "i will",
            "your" : "my",
            "yours" : "mine",
            "you" : "me",
            "me" : "you"
             }

#create chabot
chatbot=Chat(pairs,reflections)

#chat interface
def chatbot_interface():
       print("Hi ! I am Olivia , How can I assist you today?")
       while True:
           user_input=input("You : ").lower()
           if user_input=="quit":
               print("Olivia : Goodbye ! HAVE A GREAT DAY")
               break
           response =  chatbot.respond(user_input)
           print(f"Olivia : {response}")       
        
if __name__ == "__main__":
    chatbot_interface()
        
```
